# About The Dataset
This data is all about the sales data of mobile phones of various companies. Here we are predicting the price range of different mobile manufactured by various companies by finding out some relation between features of a mobile phone(eg:- RAM,Internal Memory etc)

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Download training data
df_training = pd.read_csv("ADMN5015 CLASSIFIER TRAININGDATA_NIDIYA.csv")
df_training
```

```python
# Make sure data types are correct (no objects)

df_training.dtypes
```

Here the data type of every column is integer or float so we are good to go. 

```python
# Pop label

target = df_training.pop('price_range')
```

```python
# Define features

numeric_feature_names = ['battery_power', 'blue', 'clock_speed', 'dual_sim','fc','four_g','int_memory','m_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_h','sc_w','talk_time','three_g','touch_screen','wifi']
numeric_features = df_training[numeric_feature_names]
numeric_features.head()
```

```python
# Convert data to Tensorflow tensor

numeric_features = tf.convert_to_tensor(numeric_features)
numeric_features
```

```python
# Normalize data

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(numeric_features)
```

```python
# Define model

def get_basic_model():
  model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(4) # Possible labels 0, 1 ,2 or 3
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                # loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
# Train model

BATCH_SIZE = 2002

model = get_basic_model()
model.fit(numeric_features, target, epochs=700, batch_size=BATCH_SIZE)
```

Here I put Batch size as 2002 and Epoch as 700. In the initial steps we can see that loss: 1.6027 - accuracy: 0.2280. When we kept on moving around Epoch = 500 it becomes loss: 0.0824 - accuracy: 0.9945, and when Epoch = 700, we will gwt something like this, loss: 0.0403 - accuracy: 1.0000. In this case the value of loss is much smaller than the intial one and we get an accuracy of 100% in the last one while it was 22% in the initial one.
