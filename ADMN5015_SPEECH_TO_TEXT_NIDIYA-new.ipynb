```python
 
from google.cloud import storage


from google.cloud import speech


import os

import pprint

# Create environment variables

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "admn5015-speech-to-text-bd41f6ae7e1e.json"



def list_blobs(bucket_name):

    #"""Lists all the blobs in the bucket."""

    storage_client = storage.Client()


    # Note: Client.list_blobs requires at least package version 1.17.0.

    blobs = storage_client.list_blobs(bucket_name)

    for blob in blobs:

      print(blob.name)

def extract_audio_gcs(gcs_uri):

    """Asynchronously transcribes the audio file specified by the gcs_uri."""



    client = speech.SpeechClient()


    audio = speech.RecognitionAudio(uri=gcs_uri)

    config = speech.RecognitionConfig(

        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,

        audio_channel_count=2,

        language_code="en-US",

   )



    operation = client.long_running_recognize(config=config, audio=audio)



    print("Waiting for operation to complete...")

    response = operation.result(timeout=500)



    return response


def main():

    bucket_name = "the-great-indian-bucket"

    list_blobs(bucket_name=bucket_name)

    # Extract audio from file

    gcs_uri = "gs://" + bucket_name + "/" + "SPEECH TO TEXT AUDIO _NIDIYA.wav"

    print("\nThis is the uri: {}".format(gcs_uri))



    response = extract_audio_gcs(gcs_uri=gcs_uri)
    pprint.pprint(response)
    
if __name__ == "__main__":

    main()
    
```

```python

from urllib import response 
from google.cloud import language_v1


class Sentiment(object):
    def _init_(self) -> None:
        pass
    
    def analyze_sentiment(self,text_content):
        
        client = language_v1.LanguageServiceClient()
        
        document = language_v1.Document(
            content=text_content,type_ = language_v1.Document.Type.PLAIN_TEXT
        )
        
        response = client.analyze_sentiment(
            request = {"document":document}
        ).document_sentiment
        
        return response
```
